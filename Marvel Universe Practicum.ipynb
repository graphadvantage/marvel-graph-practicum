```python
#Get all the Marvel Characters - 1484

#!pip install pandas
import time
from IPython.display import display, HTML
import requests
import json
import time
import pandas as pd
from pandas import DataFrame
import hashlib

public_key = "YOUR_PUBLIC_KEY"
private_key = "YOUR_PRIVATE_KEY"

limit = 100
offset = 0
#iterations = 1 
iterations = 15 #Holy Iterator! That's a lot of characters!
df = DataFrame()

for i in range(iterations):
    ts = str(int(time.time()))
    marvel_key = (ts + private_key + public_key).encode("UTF-8")
    #print(marvel_key)
    hash = hashlib.md5(marvel_key).hexdigest()
    request_url = 'http://gateway.marvel.com/v1/public/characters?ts=%s&apikey=%s&hash=%s&limit=%s&offset=%s' % (ts, public_key, hash, limit, offset)
    #print(request_url)
    r = requests.get(request_url)
    data = json.dumps(r.json())
    data_json = json.loads(data)["data"]["results"]
    df = df.append(data_json,ignore_index=True)
    display(df.tail(n=1))
    offset = offset + limit
    time.sleep(2)
    
#Some formatting to make a loadable CSV    

for col in ["name"]: 
    df[col] = df[col].astype(str)

for col in ["thumbnail"]: 
    df[col] = df[col].apply(lambda x: x.get('path')) + '/standard_fantastic.' + df[col].apply(lambda x: x.get('extension'))

for col in ["description"]:
    df[col] = df[col].astype(str)

for col in ["series","comics","events","stories"]:
    df[col] = df[col].map(lambda x: x["available"])
    
ordered_df = df[["id","name","thumbnail","description","series","comics","events","stories"]]
ordered_df = ordered_df.sort_values("id", ascending=False)
#display(ordered_df.tail(n=1))

ordered_df.to_csv("[PATH_TO_DIRECTORY]/marvel-graph-practicum/marvel_characters.csv",encoding='utf-8', index=False, quoting=1)

print("Done!")

```
